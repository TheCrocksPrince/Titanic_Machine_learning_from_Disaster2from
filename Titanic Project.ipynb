### Missing Values Embarked

```python
df_train[df_train['Embarked'].isnull()]
```

```python
df_train['Embarked'].value_counts()
```

```python
modus_embarked = df_train['Embarked'].mode()[0]
df_train['Embarked'] = df_train['Embarked'].fillna(modus_embarked)
```

### Missing Values Age

```python
mean_age = df_train['Age'].mean()
df_train['Age'] = df_train['Age'].fillna(mean_age)
```

```python
mean_age_test = df_test['Age'].mean()
df_test['Age'] = df_test['Age'].fillna(mean_age_test)
```

### Missing Values Fare

```python
mean_fare_test = df_test['Fare'].mean()
df_test['Fare'] = df_test['Fare'].fillna(mean_fare_test)
```

### Missing Values Cabin

Drop data cabin karna terlalu banyak missing value

```python
df_test.head()
```

```python
df_test = df_test.drop(['Cabin'], axis = 1)
```

```python
df_train = df_train.drop(['Cabin'], axis = 1)
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

## Converting Data To Numeric

### Sex Mapping

```python
sex_mapping = {"male" : 1 , "female" : 0}

df_train['Sex'] = df_train['Sex'].map(sex_mapping)
df_test['Sex'] = df_test['Sex'].map(sex_mapping)
```

```python
df_train.head()
```

### Embarked Mapping

```python
embarked = df_train[['Embarked']]
dummies = pd.get_dummies(embarked)

dummies.head()
```

```python
df_train = pd.merge(df_train.reset_index(), dummies.reset_index())
df_train.head()
```

```python
df_train = df_train.drop(['index','Embarked'], axis = 1)
```

```python
embarked_test = df_test[['Embarked']]
dummies_test = pd.get_dummies(embarked_test)

dummies_test.head()
```

```python
df_test = pd.merge(df_test.reset_index(), dummies_test.reset_index())
df_test.head()
```

```python
df_test = df_test.drop(['index','Embarked'], axis = 1)
```

### Extract Data Title From Name and Convert Tittle

```python
df_train['Title'] = df_train['Name'].str.extract(' ([A-Za-z]+)\.' , expand = False)
```

```python
df_test['Title'] = df_test['Name'].str.extract(' ([A-Za-z]+)\.' , expand = False)
```

```python
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2, 
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }

df_train['Title'] = df_train['Title'].map(title_mapping)
```

```python
df_test['Title'] = df_test['Name'].str.extract(' ([A-Za-z]+)\.', expand = False)
df_test['Title'] = df_test['Title'].map(title_mapping)
```

## Modelling

```python
# Importing Classifier Modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
train_data = df_train.drop("Survived",axis = 1 )
target = df_train['Survived']
```

```python
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)

clf = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
score.mean()
```

```python
k_fold = KFold(n_splits=5 , shuffle=True, random_state=0)

clf = RandomForestClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
score.mean()
```

```python
k_fold = KFold(n_splits=5 , shuffle=True, random_state=0)

clf = GaussianNB()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
score.mean()
```

```python
k_fold = KFold(n_splits=5 , shuffle=True, random_state=0)

clf = KNeighborsClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
score.mean()
```

```python
k_fold = KFold(n_splits=5 , shuffle=True, random_state=0)

clf = SVC()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
score.mean()
```
